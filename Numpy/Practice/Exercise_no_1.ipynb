# Create an array of student marks. Find the average, max, mean, percentage, Failures and Toppers.

```python
totalMarks = 1100
marks = np.random.randint(1, 1101, 100)
marks
```

Average

```python
avg = np.average(marks)
avg
```

Max

```python
maximun = np.max(marks)
maximun
```

Min

```python
minimum = np.min(marks)
minimum
```

Mean

```python
mean = np.mean(marks)
mean
```

Percentage

```python
percentage = marks * 100/ totalMarks
percentage = np.round(percentage, 2)
percentage
```

Failures

```python
failures = percentage <= 40 
failures = marks[failures]
failures
```

Toppers

```python
toppers = percentage >= 80
toppers = marks[toppers]
toppers
```

```python
dic = { 
    "Total Marks": totalMarks,
    "Average" : avg,
    "Maximum" : max,
    "Minimum" : min,
    "Mean" : mean,
    "Percentage" : percentage,
    "Failure" : failures,
    "Topper" : toppers 
}
```

```python
for keys,values in dic.items():
    print(f"{keys} = {values}")
```
