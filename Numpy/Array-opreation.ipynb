## Array Opreation

Arthematic Opreation

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([6,7,8,9,10])
```

```python
arr1 + arr2
```

```python
arr1 - arr2
```

```python
arr1 * arr2
```

```python
arr1 / arr2
```

```python
arr1 // arr2
```

```python
arr1 ** arr2
```

```python
arr1 % arr2
```

```python
arr = np.arange(1,10).reshape(3,3)
ARR = np.arange(11,20).reshape(3,3)
```

```python
arr + ARR
```

Matrix Opreation

```python
mat1 = np.arange(1,5).reshape(2,2)
mat2 = np.arange(5,9).reshape(2,2)
```

```python
mat1
```

```python
mat2
```

```python
# Multipilication in Matrix
mat1 @ mat2
np.dot(mat1,mat2)
```

```python
# Transpose in Matrix
mat1.T
```

Advanced Array Manipulation

```python
# Stacking Array
```

```python
A = np.arange(1,6)
B = np.arange(6,11)
```

```python
A
```

```python
B
```

```python
# Vertical Stack
np.vstack((A,B))
```

```python
# Horizontal Stack
np.hstack((A,B))
```

```python
# Colume Stack
np.column_stack((A,B))
```

```python
# Splitting Array
```

```python
arr = np.arange(1,17).reshape(4,4)
arr
```

```python
# Horizontal Split
np.hsplit(arr,2)
```

```python
# Vertical Split
np.vsplit(arr,2)
```
