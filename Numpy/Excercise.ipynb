## Valid Sudoku

```python
sudoku = np.array([
      [5, 3, 4, 6, 7, 8, 9, 1, 2],
      [6, 7, 2, 1, 9, 5, 3, 4, 8],
      [1, 9, 8, 3, 4, 2, 5, 6, 7],

      [8, 5, 9, 7, 6, 1, 4, 2, 3],
      [4, 2, 6, 8, 5, 3, 7, 9, 1],
      [7, 1, 3, 9, 2, 4, 8, 5, 6],

      [9, 6, 1, 5, 3, 7, 2, 8, 4],
      [2, 8, 7, 4, 1, 9, 6, 3, 5],
      [3, 4, 5, 2, 8, 6, 1, 7, 9]
    ])
```

```python
#For Rows
rows = np.sum(sudoku, axis = 1)
for i in rows:
    if i != 45:
        print("Sudoku is'nt valid")
        break
else:
    print("Sudoku is valid for rows")
```

```python
#For Columns
columns = np.sum(sudoku, axis = 0)
for i in columns:
    if i != 45:
        print("Sudoku is'nt valid")
        break
else:
    print("Sudoku is valid for columns")
```

```python
# For Blocks
for i in range(0,9,3):
    for j in range(0,9,3):
        n = sudoku[ i : i + 3 , j : j + 3 ]
        print(np.sum(n))
```

## Generals Question

```python
# Columns: [Age, Math's marks, Science marks]
data = np.array([
    [16,67,98],   # Student  01
    [15,86,56],   # Student  02
    [16,78,25],   # Student  03
    [17,65,70],   # Student  04
    [14,90,93]    # Student  05
])
```

```python
# Gete the shape of the matrix
shape = data.shape
shape
```

```python
# Find the avg age of student
age = data[0:,0]
np.average(age)
```

```python
# Extract Math marks of all student
math_mark = data[:,1]
math_mark
```

```python
# Find the Highest Science marks
sci_marks = data[:,2]
np.max(sci_marks)
```

```python
# Get detail of the student who scored more than 90
marks = (math_mark + sci_marks) * 100/200
marks[marks >= 90]
```

```python
# Increse math marks of all student by 5
new_maths_marks = math_mark + 5 * 100/200
new_maths_marks
```

```python
# Find How many Student are Younger than 16
is_young = age[age > 16]
len(is_young)
```

```python
# Calculate the avg marks in each subject(Column-Wise Mean)
np.mean(data[:,1:],axis = 1)
```

```python
# Get dat of student who scored at least 80 in both subject
marks[marks >= 80]
```
