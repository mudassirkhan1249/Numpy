```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 2*np.pi, 500)
```

```python
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)
```

```python
plt.figure(figsize=(10,8))

plt.subplot(3,1,1)
plt.plot(x, sin_x)
plt.title("Sine Wave")

plt.subplot(3,1,2)
plt.plot(x, cos_x)
plt.title("Cosine Wave")

plt.subplot(3,1,3)
plt.plot(x, tan_x)
plt.ylim(-10,10)
plt.title("Tangent Wave")

plt.tight_layout()
plt.show()
```

```python
analysis = {
    "sin_max": np.max(sin_x),
    "sin_min": np.min(sin_x),
    "cos_mean": np.mean(cos_x),
    "tan_std": np.std(tan_x)
}

analysis
```
